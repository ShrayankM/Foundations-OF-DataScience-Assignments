```
import numpy as np
import pandas as pd
import seaborn as sns
```

#Uniform Distribution

```
np.random.uniform?
```

```
uniform_samples = (np.random.uniform(0, 1, 100000))
```

```
sns.distplot(uniform_samples, bins=10, hist=True, kde=False);
```

```
samples = [np.sum(np.random.uniform(0, 1, 100)) for x in range(10000)]
```

```
sns.distplot?
```

```
sns.distplot(samples, bins=20, hist=True, kde=True);
```

#Normal Distribution

```
normal_samples = (np.random.normal(0, 1, 100000))
```

```
sns.distplot(normal_samples, bins=20, hist=True, kde=True);
```

```
samples = [np.sum(np.random.normal(0, 1, 100)) for x in range(10000)]
```

```
sns.distplot(samples, bins=20, hist=True, kde=True);
```

#Discrete Distribution

```
discrete_samples = np.random.randint(1, 101, 100)
```

```
sns.distplot(discrete_samples, bins = 20, hist = True, kde = True)
```

```
samples = [np.sum(np.random.randint(1, 101, 100)) for x in range(100000)] 
```

```
sns.distplot(samples, bins=20, hist=True, kde=True);
```
