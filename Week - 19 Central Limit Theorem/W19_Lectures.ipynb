#10 Die Throws example

```
import plotly.express as ply
import matplotlib.pyplot as plt
```

```
MU = np.mean([x for x in range(1, 7)])
SD = np.std([x for x in range(1, 7)])
```

```
print("Mean > " + str(MU))
print("Standard Deviation > " + str(SD))
```

```
samples = [np.sum([np.random.randint(1, 7) for n in range(10)]) for k in range(10000)]
```

```
MU = np.mean(samples)
SD = np.std(samples)
```

```
print("Mean > " + str(MU))
print("Standard Deviation > " + str(SD))
```

```
normal_samples = (samples - MU)/SD
```

```
sns.set_color_codes('bright')
```

```
plt.legend?
```

```
plt.figure(figsize = (20, 8))
sns.distplot(samples, bins=20, hist=True, kde=True, label = 'Original Distribution');
sns.distplot(normal_samples, bins=20, hist=True, kde=True, label = 'Normalized Distribution');

plt.legend(loc = 'best', fontsize = 18, fancybox = True, shadow = True)
```

```
import plotly.figure_factory as ff
group_labels = ['Originalplot', 'Normalplot'] # name of the dataset
hist_data = [samples, normal_samples]

fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```

### Chance of getting a sum greater than 40 after 10 die throws

```
round(round(stats.norm.cdf((59.5 - MU)/SD), 3) - round(stats.norm.cdf((40.5 - MU)/SD), 3), 3)
```
