### Changing Mean and n

```
def change_mu(x):
   samples = [np.var(np.random.normal(x, 1, 3)) for n in range(10000)] 
   plt.figure(figsize = (10, 7))
   sns.distplot(samples, bins = 30, hist = True, kde = False);
```

```
for x in range(0, 5):
    change_mu(x)
```

```
def change_n(x):
   samples = [np.var(np.random.normal(0, 1, x)) for n in range(10000)] 
   plt.figure(figsize = (10, 7))
   sns.distplot(samples, bins = 30, hist = True, kde = False);
```

```
for n in range(3, 7):
    change_n(n)
```

```
sample_var_x = []
```

```
def change_n(x):
   samples = [np.var(np.random.normal(0, 1, x)) for n in range(10000)] 
   sample_var_x.append(np.mean(samples))
```

```
for x in range(3, 50):
    change_n(x)
```

```
plt.figure(figsize = (10, 7))
sns.lineplot(x = [x for x in range(3, 50)], y = sample_var_x);
sns.lineplot(x = [x for x in range(3, 50)], y = [1 for x in range(3, 50)]);
```

#Chi Square Distribution

```
def generate_random(i = 0, j = 1, n = 1):
    t = np.random.normal(i, j, n)
    return t * t
```

```
def square_distribution(n):
    samples = [np.sum(generate_random(0, 1, n)) for x in range(1000)]
    plt.figure(figsize = (10, 7))
    sns.distplot(samples, bins = 30, hist = True, kde = True);
```

```
for n in range(1, 10):
    square_distribution(n)
```
