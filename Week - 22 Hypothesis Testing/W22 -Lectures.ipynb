```python
import numpy as np
import pandas as pd
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def area_plot(z_score, double, color):
    x_min = -4
    x_max = 10

    mean = 0
    std = 1

    x = np.linspace(x_min, x_max, 100)
    y = norm.pdf(x, mean, std)

    ax = plt.gca()
    plt.xlim(x_min, x_max)
    plt.ylim(0.0, 0.7)

    plt.plot(x, y);

    if z_score < 0:
        x1 = np.linspace(z_score, x_min, 100)
    else:
        x1 = np.linspace(z_score, x_max, 100)
    y1 = norm.pdf(x1, mean, std)

    if double == 2:
        if z_score < 0:
            x2 = np.linspace(z_score * -1, x_max, 100)
        else:
            x2 = np.linspace(z_score * -1, x_min, 100)
        y2 = norm.pdf(x2, mean, std)
        plt.fill_between(x2, y2, color = color, alpha = 1);
    

    plt.fill_between(x1, y1, color = color, alpha = 1);

    plt.show()
    plt.close()
```

# Chips Dataset
A company selling chips claims that each packet has 200 grams of chips (as mentioned on packet). You are secptic of their belief that on average each packet does not contain 200 grams of chips. Based on past data you know that the standard deviation is 10 grams. Prove your claim ?

```python
chips_data = [193, 212, 174, 200, 195, 195, 194, 198, 181, 203]
mu_p = 200
sigma_p = 10
n = 10
```

```python
#Calculate sample mean
mu_s = np.mean(chips_data)

#Calculate z statistics
z_score = (mu_s - mu_p)/(sigma_p/math.sqrt(n))

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

p_value = 2 * (1.0 - norm.cdf(np.abs(z_score))) * 100
# p_value = float("{:.4f}".format(p_value))
p_value = "{:.4f}%".format(p_value)

print(z_score,  z_critical)

if (np.abs(z_score) > z_critical):
    print("Reject the null Hypothesis: p-value = ", p_value, " alpha = ", alpha)
else:
    print("Not enough evidence to reject the null Hypothesis: p-value = ", p_value, " alpha = ", alpha)
```

```python
area_plot(z_score, 2, "orange")
```

# User Rating System
You have developed a new dialog system and done a user study. You claim that the average rating given by the users is greater than 4 on a scale of 1 to 5. How do you prove this to you critiques ? (std given 0.5)


```python
user_data = [4, 3, 5, 4, 5, 3, 5, 5, 4, 2, 4, 5, 5, 4, 4, 5, 4, 5, 4, 5]
n = len(user_data)
mu_p = 4
sigma_p = 0.5
```

```python
#Calculate sample mean
mu_s = np.mean(user_data)

#Calculate z_score
z_score = (mu_s - mu_p)/(sigma_p/math.sqrt(n))

alpha = 0.05

z_critical = norm.ppf(1.0 - alpha)

p_value = (1.0 - norm.cdf(np.abs(z_score))) * 100
p_value = "{:.4f}%".format(p_value)

print(z_score, z_critical)

if (z_score > z_critical):
    print("Reject the null Hypothesis: p-value = ", str(p_value), "alpha = ", alpha)
else:
    print("Not enough evidence to reject the null Hypothesis: p-value = ", p_value, " alpha = ", alpha)
```

```python
area_plot(z_score, 1, "red")
```
