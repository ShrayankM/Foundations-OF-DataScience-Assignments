# Student's T distribution

```
finalsamples = []
mu = 10
n = 10
```

```
def create_samples(i = 10, j = 30):
    samples = []
    while i <= j:
        tsamples = [np.random.normal(0, 20, i) for x in range(10000)]
        i = i + 5
        tmean = np.array([np.mean(tsamples[k]) for k in range(10000)])
        tvar  = np.array([np.var(tsamples[k]) for k in range(10000)])
        fsamples =  math.sqrt(i) * ((tmean)/tvar)
        samples.append(fsamples)
    return samples
```

```
finalsamples = create_samples()
```
